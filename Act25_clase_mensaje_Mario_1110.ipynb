## el contexto de la Programación Orientada a Objetos (POO) que estamos practicando, el "Mensaje" tiene dos significados: uno técnico y otro práctico dentro de tu código.

![texto del vínculo](https://raw.githubusercontent.com/MarioMarentes128/Act25-Clase-mensaje-Mario-1110/refs/heads/main/infoclasemensaje.png)

```python
class Mensaje:
    def __init__(self, emisor, receptor, contenido):
        # Inicialización de los 3 atributos
        self.emisor = emisor
        self.receptor = receptor
        self.contenido = contenido


    # Función 1: Lectura de datos
    def leer_datos(self):
        print(f"\n--- Configurando nuevo mensaje ---")
        self.emisor = input("Emisor: ")
        self.receptor = input("Receptor: ")
        self.contenido = input("Mensaje: ")

    # Función 2: Muestra de datos
    def mostrardatos(self):
        print("\n📥 TICKET DE ENVÍO")
        print(f"DE: {self.emisor}")
        print(f"PARA: {self.receptor}")
        print(f"TEXTO: {self.contenido}")

    # Función 3: Confirmación de sistema
    def confirmar_envio(self):
        # Aquí procesamos una variable discreta
        codigo_secreto = "amo al profe nava"
        print(f"Sincronizando con el servidor...")
        print(f"Estado: Listo para enviar de {self.emisor} a {self.receptor}.")

# --- EJECUCIÓN ---

# Creamos los 2 objetos (mensaje1 y mensaje2)
mensaje1 = Mensaje("Sujeto A", "Sujeto B", "Secreto")
mensaje2 = Mensaje("Sujeto C", "Sujeto D", "Privado")

# Interacción con Mensaje 1
mensaje1.leer_datos()
mensaje1.confirmar_envio()
mensaje1.mostrardatos()

# Interacción con Mensaje 2
mensaje2.leer_datos()
mensaje2.confirmar_envio()
mensaje2.mostrardatos()

print("\nSistema finalizado exitosamente.")
print("Programa terminado")
print("Programa realizado por Mario NL = 53")
```
